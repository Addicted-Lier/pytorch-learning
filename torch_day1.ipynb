```python
import torch
#1d tensor
x = torch.tensor([1,2,3])

#2d tensor

y = torch.tensor([[1,2,3], [4,5,6]])
print(x)
print(y)
```

```python
a = torch.rand(3,3)
b = torch.zeros(3,3)
c = torch.ones(2,3)

print(a)
print(b)
print(c,"\n")

print("printing the shapes of each tensor\n")

print(a.shape)
print(b.shape)
print(c.shape)
```

```python
print(a+b)
print(a*b)
```

```python
#3d tensors
# 👉 Meaning:

# 2 matrices
# each has 2 rows
# each row has 2 columns

# Think:

# (depth, rows, columns)
import torch
x = torch.tensor([
    [[1,2],[3,4]],
     [[23,34],[25,27]]
    ])

print(x)
print(x.shape)


# 🎯 REAL LIFE USE (IMPORTANT)

# 3D tensors are used in:

# 📸 Images (very important for you)
# 3D = (channels, height, width)

# Example:

# RGB image → 3 channels
# 👉 shape = (3, H, W)
```

```python
x = torch.rand(3,2,2)
print(x)
print(x.shape)
```
