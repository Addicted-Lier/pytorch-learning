```python
import torch

# Create a 'Synthetic Leaf' (3 channels, 224x224 pixels)
# Channel 0 = Red, Channel 1 = Green, Channel 2 = Blue
leaf_tensor = torch.zeros(3, 224, 224, device="cuda")

# Let's make it a 'Green' leaf by filling the Green channel
leaf_tensor[1, :, :] = 1.0 

print(f"Leaf Tensor Shape: {leaf_tensor.shape}")
print(f"Device: {leaf_tensor.device}")
print("✅ The GPU is now 'holding' a digital green leaf!")
```

```python
import matplotlib.pyplot as plt

# 1. Move from GPU to CPU (The waiter brings the dish out)
# 2. Change shape from (Channels, Height, Width) to (Height, Width, Channels) 
# 3. .numpy() converts it to a format matplotlib understands
leaf_to_show = leaf_tensor.cpu().permute(1, 2, 0).numpy()

# 4. Show the image
plt.imshow(leaf_to_show)
plt.title("My First Synthetic Green Leaf")
plt.axis('off') # Hide the X and Y coordinates
plt.show()
```

```python
# Create a yellow/brown spot in the middle of our green leaf
# Red + Green = Yellow. Red + Green + a bit of Blue = Brown.
leaf_tensor[0, 100:120, 100:120] = 0.5  # Add some Red
leaf_tensor[1, 100:120, 100:120] = 0.3  # Reduce some Green

# Show it again
plt.imshow(leaf_tensor.cpu().permute(1, 2, 0).numpy())
plt.title("Leaf with a 'Disease' Spot")
plt.show()
```
